```python
import requests
from PIL import Image
import io
```

```python
image = Image.open('cat_224x224.jpg')
print("Original photo")
image
```

```python
resp = requests.post("http://127.0.0.1:8000/superresolution",
                     files={"file": open('cat_224x224.jpg','rb')})
resp
```

```python
image = Image.open(io.BytesIO(resp.content))
print("Flask service result")
image
```
